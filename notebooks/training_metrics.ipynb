# Visualization of YOLOv5 Training Metrics for Human Fall Detection

This notebook provides visual analysis of the training performance of a YOLOv5 model trained on a custom fall detection dataset. The dataset includes three classes: **Fall-Action**, **Sit**, and **Stand**.  
The model was trained for **500 epochs**, and this notebook helps evaluate:

- ðŸ“ˆ Precision, Recall, and mAP scores across epochs  
- ðŸ“‰ Training and validation losses (box, objectness, class)

Dataset preparation, training, and deployment were carried out as part of a final year project (PFA) at ENIT.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Style moderne
sns.set(style="whitegrid")

# Charger les donnÃ©es (remplace par le bon chemin si nÃ©cessaire)
df = pd.read_csv('rsultats.csv')
df.columns = df.columns.str.strip()  # Nettoyage des noms de colonnes

# Afficher les colonnes disponibles
print("Colonnes disponibles :", df.columns.tolist())

# Liste des mÃ©triques Ã  tracer
metrics = ['metrics/precision', 'metrics/recall', 'metrics/mAP_0.5', 'metrics/mAP_0.5:0.95']
losses = ['train/box_loss', 'train/obj_loss', 'train/cls_loss', 'val/box_loss', 'val/obj_loss', 'val/cls_loss']

# Tracer les courbes des mÃ©triques
plt.figure(figsize=(12, 6))
for metric in metrics:
    if metric in df.columns:
        plt.plot(df[metric], label=f"{metric} (moy: {df[metric].mean():.3f})")
plt.title("Ã‰volution des mÃ©triques de performance")
plt.xlabel("Ã‰poque")
plt.ylabel("Valeur")
plt.legend()
plt.show()

# Tracer les courbes de loss
plt.figure(figsize=(12, 6))
for loss in losses:
    if loss in df.columns:
        plt.plot(df[loss], label=f"{loss} (moy: {df[loss].mean():.3f})")
plt.title("Ã‰volution des pertes (loss)")
plt.xlabel("Ã‰poque")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.show()
```
